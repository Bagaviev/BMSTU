```python
import googlemaps
import pandas as pd
from datetime import datetime

gmaps = googlemaps.Client(key='')
page1 = gmaps.places(query='ZARA', location='55.753440,37.620522', radius='20000', language='RU')
```

```python
page1
```

#### Если результат запроса содержит более 20 точек, то вернется N страниц (не более 3-х) с результатами запроса, с параметром next_page_token, иначе - 20 строк без него

```python
page1_PD = pd.DataFrame(page1['results'])
page1_PD[['name', 'geometry', 'formatted_address']].head(70)
```

```python
page1.keys()
```

```python
next_page_token = page1['next_page_token']
```

```python
page2 = gmaps.places(query='ZARA', location='55.753440,37.620522', radius=1000, language='RU', page_token=next_page_token)
page2_PD = pd.DataFrame(page2['results'])
page2_PD[['name', 'geometry', 'formatted_address']].head(70)
```

```python
page2.keys()
```

```python
next_page_token = page2['next_page_token']
page3 = gmaps.places(query='ZARA', location='55.753440,37.620522', radius=1000, language='RU', page_token=next_page_token)
page3_PD = pd.DataFrame(page3['results'])
page3_PD[['name', 'geometry', 'formatted_address']].head(70)
```

```python
mask = ['name', 'geometry', 'formatted_address']
page_result = pd.concat([page1_PD[mask], page2_PD[mask], page3_PD[mask]])
page_result.to_csv('WORK/zara.csv', encoding='utf-8', index=False, sep=';')
```
